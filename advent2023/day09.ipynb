## Day 9: Mirage Maintenance:

[#](https://adventofcode.com/2023/day/9) We have a list of values - each line is the history of a particular value.

For each line of values:
* Make a list of differences b/w each value, and keep repeating until the differences are zero
* use that to find the next predicted value

Return the sum of the next predicted values for each line.


```python
test: str = """0 3 6 9 12 15
1 3 6 10 15 21
10 13 16 21 30 45"""

inp = utils.get_input(9, splitlines=False)
```

```python
def parse(inp=test, verbose=False):
    """returns a list of ints for each line"""
    nums = []
    for line in inp.strip().splitlines():
        nums.append([int(n) for n in re.findall(r"\d+", line)])
    return nums


nums = parse()
nums
```

```python
def step_line(line: list[int]) -> list[int]:
    return [line[i] - line[i - 1] for i in range(1, len(line))]


def predict_line(line, debug: bool = False):
    """recursively predicts the next num in sequence"""
    if len(set(line)) == 1:
        return line[-1]

    if debug:
        print(line)
    return line[-1] + predict_line(step_line(line))


for line in nums:
    print(line[:5], predict_line(line))
```

```python
def solve(inp=test, verbose: bool = False):
    nums = parse(inp)
    n = [predict_line(line) for line in nums]
    print(f"found {len(n)} next numbers for {len(nums)} lines")
    return sum(n)


assert solve(test) == 114  # example answer
solve(inp)
```

```python
lines = []
for line in parse(inp):
    n = predict_line(line)
    line.append(n)
    lines.append(line)

print(lines[0])
px.line(y=lines, log_y=0)
```
